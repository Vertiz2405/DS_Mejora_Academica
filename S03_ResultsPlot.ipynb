MUSIC Inventory (College Student version)

To obtain a score for each scale, average the values for the items in the scales as shown below.

1. Empowerment score = (item 2 + item 8 + item 12 + item 17 + item 26) / 5

2. Usefulness score = (item 3 + item 5 + item 19 + item 21 + item 23) / 5

3. Success score = (item 7 + item 10 + item 14 + item 18) / 4

4. Interest score = (item 1 + item 6 + item 9 + item 11 + item 13 + item 15) / 6

5. Caring score = (item 4 + item 16 + item 20 + item 22 + item 24 + item 25) / 6

```python
# -------------------------------
# Print dataset MOTIVATION
dfmo
```

```python
# -------------------------------
# Calcular MUSIC scores

dfmusic = pd.DataFrame()

dfmusic["Empowerment"] = ( dfmo["Q2"]+dfmo["Q8"]+dfmo["Q12"]+dfmo["Q17"]+dfmo["Q26"] )/5

dfmusic["Usefulness"]  = ( dfmo["Q3"]+dfmo["Q5"]+dfmo["Q19"]+dfmo["Q21"]+dfmo["Q23"] )/5

dfmusic["Success"]     = ( dfmo["Q7"]+dfmo["Q10"]+dfmo["Q14"]+dfmo["Q18"] )/4

dfmusic["Interest"]    = ( dfmo["Q1"]+dfmo["Q6"]+dfmo["Q9"]+dfmo["Q11"]+dfmo["Q13"]+dfmo["Q15"] )/6

dfmusic["Caring"]      = ( dfmo["Q4"]+dfmo["Q16"]+dfmo["Q20"]+dfmo["Q22"]+dfmo["Q24"]+dfmo["Q25"] )/6

# -------------------------------
# In case it is needed
#Nque = 26
#Vmin = 1*Nque
#Vmax = 6*Nque
#dfmusic["Grade"] = 100* ( (dfmo.sum(axis=1)-Vmin) / (Vmax-Vmin) )
```

```python
# -------------------------------
# Plot distribution of MUSIC scores

fig, ax = plt.subplots( figsize=(6,3), dpi=dpival )

sns.boxplot(data=dfmusic, orient="v", palette="Set3")

plt.title("MUSIC Inventory")
plt.xticks(rotation=30)
plt.ylabel("Score")
plt.ylim(0.5,6.5)
plt.show()
```

```python
# -------------------------------
# Summary
dfmusic.describe()
```

```python
# -------------------------------
# Statistical test

# Import Kruskal-Wallis test
from scipy.stats import kruskal

# Conduct the Kruskal-Wallis Test
kruskal(dfmusic["Empowerment"].values, dfmusic["Usefulness"].values, dfmusic["Success"].values, dfmusic["Interest"].values, dfmusic["Caring"].values)
```

```python
# Defining data groups
data_group1 = [7, 9, 12, 15, 21]
data_group2 = [5, 8, 14, 13, 25]
data_group3 = [6, 8, 8, 9, 5]

# Conduct the Kruskal-Wallis Test
result = kruskal(data_group1, data_group2, data_group3)

# Print the result
print(result)
```
